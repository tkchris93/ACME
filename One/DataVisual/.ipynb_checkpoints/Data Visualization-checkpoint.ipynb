```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.misc import comb
from mpl_toolkits.mplot3d import Axes3D
```

#Data Visualization

```python
%matplotlib notebook
def problemOne():
    """Order and plot the data as a horizontal bar chart."""
    
    male = np.array([179.2,160.0,177.8,178.9,178,176,172.5,165.6,170.8,
                     183.2,182.4,164,163.6,175.4,174,176.1,165.7])
    female = np.array([167.6,142.2,164.5,165.3,165,164,158,154.9,157.4,
                       168.4,168,151,151.4,164,158.9,162.1,155.2])
    country = np.array(['Austria', 'Bolivia', 'England', 'Finland', 'Germany', 'Hungary', 'Japan','North Korea', 'South Korea', 
                        'Montenegro', 'Norway', 'Peru','Sri Lanka', 'Switzerland', 'Turkey', 'U.S.', 'Vietnam'])
    male_mask = np.argsort(male)
    female_mask = np.argsort(female)
    
    plt.subplot(2,1,1)
    pos = np.arange(len(country)) + .5
    plt.barh(pos, male[male_mask], align='center')
    plt.yticks(pos, country[male_mask])
    
    plt.subplot(2,1,2)
    plt.barh(pos, female[female_mask], align='center')
    plt.yticks(pos, country[female_mask])
    plt.show()
problemOne()
```

```python
%matplotlib notebook
def problemTwo():
    """Plot some histograms with white reference lines.  Do this for 20 bins,
    10 bins, 5 bins, and 3 bins
    """
    normal_numbers = np.random.normal(size=1000)
    
    plt.subplot(2,2,1)
    plt.title("20 bins")
    plt.hist(normal_numbers, bins=20)
    plt.grid(True, color = 'w', linestyle = '-')
    
    plt.subplot(2,2,2)
    plt.title("10 bins")
    plt.hist(normal_numbers, bins=10)
    plt.grid(True, color = 'w', linestyle = '-')
    
    plt.subplot(2,2,3)
    plt.title("5 bins")
    plt.hist(normal_numbers, bins=5)
    plt.grid(True, color = 'w', linestyle = '-')
    
    plt.subplot(2,2,4)
    plt.title("3 bins")
    plt.hist(normal_numbers, bins=3)
    plt.grid(True, color = 'w', linestyle = '-')
    plt.show()
problemTwo()
```

```python
%matplotlib notebook
def problemThree():
    """Plot y = x^2 * sin(x) using 1000 data points and x in [0,100]"""
    x = np.linspace(0,100,1000)
    y = x**2 * np.sin(x)
    plt.plot(x,y)
    plt.show()
problemThree()
```

```python
%matplotlib notebook
def problemFive():
    """Plot a contour map of z = sin(x) + sin(y) where x is in [0,12*pi] and
    y is in [0,12*pi]
    """
    x = np.linspace(0,12*np.pi,400)
    y = np.linspace(0,12*np.pi,400)
    X,Y = np.meshgrid(x,y)
    Z = np.sin(X) + np.sin(Y)
    plt.contourf(X,Y,Z,cmap=plt.get_cmap('afmhot'))
    plt.show()
problemFive()
```

###Observations in Problem 6
In Problem 6 we see that just because sets of data have the same mean, variance, correlation and linear regression, it does not mean that the sets of data have similar structures at all.

```python
%matplotlib notebook
def problemSeven():
    """Change the surface to a heatmap or a contour plot.  Return a string
    of the benefits of each type of visualization.
    """
    x = np.linspace(-2*np.pi,2*np.pi,num=400)
    y = np.linspace(-2*np.pi,2*np.pi,num=400)
    X, Y = np.meshgrid(x,y)
    Z = np.exp(np.cos(np.sqrt(X**2 + Y**2)))
    plt.contourf(X,Y,Z,cmap=plt.get_cmap('Greys'))

problemSeven()    
```

```python
%matplotlib notebook
def problemEight():
    """Plot y = x^2 * sin(x) where x is in [0,100] and adjust to y limit to be
    [-10^k,10^k] for k = 0,1,2,3,4.
    """
    for k in xrange(5):
        ax = plt.subplot(5,1,k+1)
        x = np.linspace(0,100,1000)
        y = x**2 * np.sin(x)
        ax.plot(x,y)
        plt.ylim(-10**k,10**k)
    plt.show()
problemEight()
```

```python
%matplotlib notebook
def problemNine():
    """Simplify graph from problem 1."""
    
    male = np.array([179.2,160.0,177.8,178.9,178,176,172.5,165.6,170.8,
                     183.2,182.4,164,163.6,175.4,174,176.1,165.7])
    female = np.array([167.6,142.2,164.5,165.3,165,164,158,154.9,157.4,
                       168.4,168,151,151.4,164,158.9,162.1,155.2])
    country = np.array(['Austria', 'Bolivia', 'England', 'Finland', 'Germany', 'Hungary', 
                        'Japan','North Korea', 'South Korea', 'Montenegro', 'Norway', 
                        'Peru','Sri Lanka', 'Switzerland', 'Turkey', 'U.S.', 'Vietnam'])
    male_mask = np.argsort(male)
    female_mask = np.argsort(female)
      
    plt.suptitle("Heights of Males and Females by Country", fontsize=14, fontweight='bold')
    plt.subplot(2,1,1)
    plt.title("Average Male Height")
    pos = np.arange(len(country)) + .5
    plt.barh(pos, male[male_mask], align='center')
    plt.yticks(pos, country[male_mask])
    plt.xlim(0,200)
    plt.gca().set_xlabel("Height (cm)")
    plt.gca().set_ylabel("Country")
    
    plt.subplot(2,1,2)
    plt.title("Average Female Height")
    plt.barh(pos, female[female_mask], align='center')
    plt.yticks(pos, country[female_mask])
    plt.xlim(0,200)
    plt.gca().set_xlabel("Height (cm)")
    plt.gca().set_ylabel("Country")
    plt.show()
problemNine()
```

```python
%matplotlib notebook
def problemTen():
    """Plot the Bernstein polynomials for v,n in [0,3] as small multiples
    and as the cluttered version.
    """
    B = lambda n,v,x : comb(n,v)*x**v*(1-x)**(n-v)
    
    x = np.linspace(0,1,500)
    ax_cluster = plt.subplot2grid((2,4), (0,0), rowspan=2, colspan=2)
    for v in xrange(4):
        plt.plot(x,B(3,v,x))
    
    ax0 = plt.subplot2grid((2,4), (0,2))
    ax1 = plt.subplot2grid((2,4), (1,2))
    ax2 = plt.subplot2grid((2,4), (0,3))
    ax3 = plt.subplot2grid((2,4), (1,3))
    
    ax_multiples = [ax0,ax1,ax2,ax3]
    colors = ['b','g','r','c']
    for v in xrange(4):
        ax_multiples[v].axis([0,1,0,1])
        ax_multiples[v].plot(x,B(3,v,x),color = colors[v])
        ax_multiples[v].get_xaxis().set_ticks([])
        ax_multiples[v].get_yaxis().set_ticks([])
problemTen()
```
