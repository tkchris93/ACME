```python
# Importance Sampling
mean = 4
sigma = 1

h = lambda y : y > 3
f = lambda y : ss.norm(0,1).pdf(y)
g = lambda y : ss.norm(mean,sigma).pdf(y)
```

```python
IS_list = []
for n in xrange(5000,505000,5000):
    Y = np.random.normal(loc=mean,scale=sigma,size=n)
    IS = 1./n*np.sum(h(Y)*f(Y)/g(Y))
    IS_list.append(IS)
    #print "done with: " + str(n)
IS_list = np.array(IS_list)

#print "Expected: " + str(IS)
#terms_IS = h(Y)*f(Y)/g(Y)
#print "Var: " + str(np.var(terms_IS))
```

```python
# Monte Carlo
MC_list = []
for n in xrange(5000,505000,5000):
    X = np.random.normal(size=n)
    MC = 1./n*np.sum(h(X))    
    MC_list.append(MC)
    #print "done with: " + str(n)
MC_list = np.array(MC_list)
#print "Expected: " + str(MC)
#terms_MC = h(X)
#print "Var: " + str(np.var(terms_MC))
```

```python
# truth
truth = 1 - ss.norm(0,1).cdf(3)
print truth
```

```python
%matplotlib nbagg
MC_error = abs(truth - MC_list)
IS_error = abs(truth - IS_list)
x = np.arange(5000,505000,5000)
plt.plot(x,MC_error,color='r',label="Monte Carlo")
plt.plot(x,IS_error,color='b',label="Importance Sampling")
plt.title("Error of Estimations")
plt.xlabel("Number of Sample Points")
plt.legend()
plt.show()
```

```python
%%time

truth = 1 - ss.gamma(a=9,scale=0.5).cdf(10)

# h(x) = 1 if x > 10
# f(x) = G(9,0.5)
# g(x) = G(12,1)

h = lambda y : y > 10 
f = lambda y : ss.gamma(a=9,scale=0.5).pdf(y)
g = lambda y : ss.gamma(a=12,scale=1).pdf(y)


IS_list = []
for n in xrange(5000,705000,5000):
    Y = np.random.gamma(12,scale=1,size=n)
    IS = 1./n*np.sum(h(Y)*f(Y)/g(Y))
    IS_list.append(IS)
    #print "done with: " + str(n)
IS_list = np.array(IS_list)
IS1_error = abs(truth - IS_list)/truth
```

```python
%%time
# h(x) = 1 if x > 10
# f(x) = G(9,0.5)
# g(x) = N(12,3)

h = lambda y : y > 10
f = lambda y : ss.gamma(a=9,scale=0.5).pdf(y)
g = lambda y : ss.norm(loc=12,scale=3).pdf(y)

n = 10**6

IS_list = []
for n in xrange(5000,705000,5000):
    Y = np.random.normal(loc=12,scale=3,size=n)
    IS = 1./n*np.sum(h(Y)*f(Y)/g(Y))
    IS_list.append(IS)
    #print "done with: " + str(n)
IS_list = np.array(IS_list)
IS2_error = abs(truth - IS_list)/truth   
```

```python
%matplotlib notebook
x = np.arange(5000,705000,5000)
plt.plot(x,IS1_error,color='b')
plt.plot(x,IS2_error,color='r')

z1 = np.polyfit(x,IS1_error,1)
p1 = np.poly1d(z1)
z2 = np.polyfit(x,IS2_error,1)
p2 = np.poly1d(z2)

plt.plot(x,p1(x),color='b')
plt.plot(x,p2(x),color='r')
plt.show()
```
