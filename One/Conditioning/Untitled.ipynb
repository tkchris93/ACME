```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
w_coeffs = np.array([1, -210, 20615, -1256850, 53327946, -1672280820,
                40171771630, -756111184500, 11310276995381,
                    -135585182899530, 1307535010540395,
                    -10142299865511450, 63030812099294896,
                    -311333643161390640, 1206647803780373360,
                    -3599979517947607200, 8037811822645051776,
                    -12870931245150988800, 13803759753640704000,
                    -8752948036761600000, 2432902008176640000])
```

```python
wilk = np.poly1d(w_coeffs)
```

```python
#perturb = np.random.normal(1,1e-10,size=21)
#perturbed_coeffs = w_coeffs*perturb
#perturbed_roots = np.roots(np.poly1d(perturbed_coeffs))
```

```python
perturb = np.zeros(21)
perturb[1] = 1e-7
perturbed_coeffs = w_coeffs - perturb
```

```python
pwilk = np.poly1d(perturbed_coeffs)
```

```python
perturbed_roots = np.roots(pwilk)
perturbed_roots = np.sort(perturbed_roots)
```

```python
x = np.linspace(-1,21,500)
w = wilk(x)
pw = pwilk(x)
```

```python
%matplotlib notebook
plt.subplot(121)
plt.plot(x,w,color='b',label="Wilkson Polynomial")
plt.plot(x,pw,color='r',label="Perturbed Wilkson Polynomial")
plt.axis([-1,21,-4e13,4e13])

plt.subplot(122)
plt.scatter(np.arange(1,21),np.zeros_like(np.arange(1,21)),color='b')
plt.scatter(np.real(perturbed_roots), np.imag(perturbed_roots), 
            marker='x',color='r')
```
